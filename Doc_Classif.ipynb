```python

import pandas as pd
from collections import Counter, defaultdict

dflist = []
dflist2 = []
#Read in token and document 
for x in range(1,6):
    tempdf = pd.read_csv("/Users/carriehaykellar/Documents/CIB_Topc_Modelling/CIB_Topc_Modellingcibs_corpus50_" + str(x) + ".csv")
    dflist.append(tempdf)
    tempdf = pd.read_csv("/Users/carriehaykellar/Documents/CIB_Topc_Modelling/CIB_Topc_Modellingcibs_top_words50_" + str(x) + ".csv")
    dflist2.append(tempdf)
pd.set_option("display.max_rows", None, "display.max_columns", None)

mapped = []
mapped_scores = []

for x in range(0, len(dflist) -1): 
    #Merge tokens and collapse based on number of observations
    dfa = dflist2[x]
    dfb = dflist2[x + 1]
    word_merge = dfa.merge(dfb,on='Word', how='left')
    series = word_merge.groupby(["Topic_x","Topic_y"]).size().sort_values(ascending=False)

    #One-to-One Mapping
    one_to_one = defaultdict(list)
    for k,v in series.items():
        if k[0] not in one_to_one.keys() and k[1] not in [a for v in one_to_one.values() for a in v]:
            one_to_one[k[0]].append(k[1])

    #One-to-Many Mapping
    one_to_many = defaultdict(list)
    for k,v in series.items():
        if v > 5:
            one_to_many[k[0]].append(k[1])

    mapped.append(one_to_many)


    #For each dataframe select the column names with the top 2 highest probabilities and concatenate with document
    #ids and then merge on ids
    work_list = []
    for x in dflist:
        max_values = x.drop([x.columns[0],x.columns[1],x.columns[2]],axis=1).idxmax(axis=1)
        max_values2 = x.drop([x.columns[0],x.columns[1],x.columns[2]],axis=1).T.apply(lambda x: x.nlargest(2).idxmin())
        doc_ids = x[[x.columns[1]]]
        work_list.append(pd.concat([doc_ids,max_values,max_values2],axis=1))
    
    temp_df = work_list[0].merge(work_list[1],on=work_list[0].columns[0])

    #Check for given topic number, ex. '2' in run 1 with mapping 16 in run 2
    #True positive: row in above table contains '2' as most likely in run 1 and '16' as first or second most likely in run 2
    #False negative: row contains '2' as most likely in run 1 and does not contain '16' in either column for run 2
    #False positive: '16' appears as either first or second most likely for run 2 but does not appear most likely for run 1

    scores = {}
    for topic1 in list(one_to_many):
        tp = 0
        fp = 0
        fn = 0
        t1 = str(topic1)
        if len(one_to_many[topic1]) != 0:
            for t in one_to_many[topic1]:
                num_docs = 0
                t2 = str(int(t))
                for index,row in temp_df.iterrows():
                    if row[temp_df.columns[1]] == t1:
                        num_docs += 1
                    if row[temp_df.columns[1]] == t1 and (row[temp_df.columns[3]] == t2 or row[temp_df.columns[4]] == t2):
                        tp += 1
                    elif row[temp_df.columns[1]] == t1 and not (row[temp_df.columns[3]] == t2 or row[temp_df.columns[4]] == t2):
                        fn += 1
                    elif (row[temp_df.columns[3]] == t2 or row[temp_df.columns[4]] == t2) and not row[temp_df.columns[1]] == t1:
                        if int(row[temp_df.columns[3]]) not in one_to_many[int(row[temp_df.columns[1]])] and int(row[temp_df.columns[4]]) not in one_to_many[int(row[temp_df.columns[1]])]:
                            fp += 1
        

        if tp+fp == 0:
            precision = 0
        else:
            precision = tp/(tp+fp)

        if tp+fn == 0:
            recall = 0
        else:
            recall = tp/(tp+fn)

        scores[topic1] = (num_docs, precision, recall)

    mapped_scores.append(scores)
```
